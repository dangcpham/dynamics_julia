```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rc('text', usetex=True)
plt.rc('font', size=14)
plt.rc('legend', fontsize=13)
```

```python
files = {"Modified Euler": "output/pendulum_mod_euler.out", 
         "Leapfrog": "output/pendulum_leapfrog.out", 
         "SS3": "output/pendulum_SS3.out"}

fig, axes = plt.subplots(ncols=3, sharey=True, figsize=(20,5))
for i, val in enumerate(files.items()):
    integrator_name, f = val
    
    p,q = np.genfromtxt(f, unpack=True)
    axes[i].plot(np.mod(p - np.pi, 2*np.pi) - np.pi,np.mod(q, 2*np.pi), 
        "k.", markersize=0.025)
    
    axes[i].set_title(integrator_name)
    axes[i].set_xlim(-np.pi, np.pi)
    axes[i].set_ylim(0, 2*np.pi)
    axes[i].set_xlabel("p")
plt.subplots_adjust(wspace=0)
axes[0].set_ylabel("q");
```
